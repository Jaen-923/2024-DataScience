```python
import numpy as np
X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

X_2 = range(100)
Y_2 = [np.sin(value) for value in X_2]

plt.plot(X_1, Y_1)
plt.plot(X_2, Y_2)
plt.show() # 스크립트 파일(.py 파일)에서 그래프를 출력할 때 / 대화형 개발 환경에서 그래프가 자동으로 출력되지 않을 때
```

```python
fig, ax = plt.subplots() # s 붙여야한다.

X_1 = range(100)
Y_1 = [np.cos(value) for value in X_1]

ax.plot(X_1, Y_1)
ax.set(title='cos graph', xlabel='X', ylabel='Y')

plt.show() # 그래프 이상, 그래프가 이산적인 데이터들의 집합이여서
```

```python
# figure 생성후 axes 추가(add_subplot)
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c='r')
ax_2.plot(X_2, Y_1, c='b')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 1, 100)
y_1 = np.sin(x)
y_2 = np.cos(x)
y_3 = np.tan(x)
y_4 = np.exp(x)

fig, ax = plt.subplots(2, 2)

ax[0, 0].plot(x, y_1)
ax[0][1].plot(x, y_2)
ax[1][0].plot(x, y_3)
ax[1, 1].plot(x, y_4)

plt.show()
```

```python
ax1 = plt.subplot(321)
plt.plot(x, y_1)
ax2 = plt.subplot(322)
plt.plot(x, y_1)
ax3 = plt.subplot(312)
plt.plot(x, y_1)
ax4 = plt.subplot(325)
plt.plot(x, y_1)
ax5 = plt.subplot(326)
plt.plot(x, y_1)

plt.show()
```

```python
X_1 = range(100)
Y_1 = [value for value in X_1]

X_2 = range(100)
Y_2 = [value+100 for value in X_2]

plt.plot(X_1, Y_1, c='r')
plt.plot(X_2, Y_2, c='b')

plt.show()
```

```python
plt.plot(X_1, Y_1, c='b', linestyle='dashed')
plt.plot(X_2, Y_2, c='r', ls='dotted')

plt.show()
```

```python
fig = plt.figure()
fig.set_size_inches(10, 10)

ax_1 = fig.add_subplot(1, 2, 1)
ax_2 = fig.add_subplot(1, 2, 2)

ax_1.plot(X_1, Y_1, c='b')
ax_1.set_title('Figure 1')
ax_2.plot(X_1, Y_1, c='r')
ax_2.set_title('Figure 2')

plt.show()
```

```python
plt.plot(X_1, Y_1, color='b', linestyle='dashed', label='line_1')
plt.plot(X_2, Y_2, color='g', linestyle='dotted', label='line_2')

plt.legend(shadow=True, fancybox=False, loc='best')

plt.title('$y=ax+b$')
plt.xlabel('$x_line$')
plt.ylabel('y_line')
plt.show()
```

```python
data = [[5., 25., 50., 20.],
        [4., 23., 51., 17],
        [6., 22., 52., 19]]
X = np.arange(0, 8, 2) # [0, 2, 4, 6]
plt.bar(X+0.00, data[0], color='b', width=0.50)
plt.bar(X+0.50, data[1], color='g', width=0.50)
plt.bar(X+1.00, data[2], color='r', width=0.50)

plt.xticks(X+0.50, ("A", "B", "C", "D"))

plt.show()
```

```python
color_list = ['b', 'g', 'r']
data_label = ["A", "B", "C"] # legend에 쓸 라벨 리스트
data = np.array([[5., 5., 5., 5.],
                 [4., 23., 51., 17.],
                 [6., 22., 52., 19.]])
X = np.arange(data.shape[1])
for i in range(3):
  plt.bar(X, data[i], bottom = np.sum(data[:i], axis=0), color = color_list[i], label = data_label[i])
plt.legend()
plt.show()
```
