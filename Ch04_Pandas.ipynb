```python
import pandas as pd
import numpy as np
ipl_data = {'Team' : ['Riders', 'Riders', 'Devils', 'Devils', 'Kings', 'kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
            'Rank' : [1, 2, 2, 3, 3, 4, 1, 1, 2, 4, 1, 2],
            'Year' : [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
            'Points' : [876, 789, 863, 673, 741, 812, 756, 788, 694, 701, 804, 690]}
df = pd.DataFrame(ipl_data)
df
```

```python
df.groupby('Team')['Points'].sum() # 분할-적용-결합의 과정
```

```python
multi_groupby = df.groupby(['Team', 'Year'])['Points'].sum()
multi_groupby # 계층구조, level0, level1, 팀별 연도별 점수 총합이 나온다.
```

```python
multi_groupby.index
```

```python
multi_groupby["Devils":"Riders"]
```

```python
multi_groupby.unstack()
```

```python
triple_groupby = df.groupby(['Team', 'Year', 'Rank'])['Points'].sum()
triple_groupby
```

```python
triple_groupby.index
```

```python
triple_groupby.unstack()
```

```python
multi_groupby.swaplevel().sort_index() # sort_index : 첫 번째 인덱스를 기준으로 데이터 재정렬
```

```python
multi_groupby.groupby(level=0).sum()
```

```python
multi_groupby.groupby(level=1).sum()
```

```python
grouped = df.groupby('Team')
grouped.get_group('Riders') # 키값을 기준으로 분할된 데이터 프레임 객체 확인
```

```python
grouped.agg(max) # 각 항목별로 최댓값을 나열, 관련이 없다.
```

```python
grouped.agg(np.mean)
```

```python
grouped.transform(max)
```

```python
score = lambda x : (x - x.mean())/x.std()
grouped.transform(score) # kings 데이터는 하나이므로 NaN값
```

```python
df.groupby('Team').filter(lambda x : len(x) >= 3) # 팀 항목이 3개이상인 데이터 검색
```

```python
df.groupby('Team').filter(lambda x : x['Points'].max() > 800) # 팀별 포인트 최댓값이 800 초과인 팀 검색
```

```python
import pandas as pd
raw_data = {'subject_id' : ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
            'test_score' : [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
df_left = pd.DataFrame(raw_data)
df_left
```

```python
raw_data = {'subject_id' : ['4', '5', '6', '7', '8'],
            'first_name' : ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
            'last_name' : ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}
df_right = pd.DataFrame(raw_data)
df_right
```

```python
# 내부 병합
pd.merge(left=df_left, right=df_right, how="inner", on='subject_id')
```

```python
pd.merge(df_left, df_right, how="left", on='subject_id')
```

```python
pd.merge(df_left, df_right, how='right', on='subject_id')
```

```python
pd.merge(df_left, df_right, on='subject_id', how='outer')
```

```python
df_left.index = df_left.subject_id
del df_left['subject_id']
df_right.index = df_right.subject_id
del df_right['subject_id']
pd.merge(df_left, df_right, on='subject_id', how='inner')
```
